```python
import pandas as pd
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from tqdm import tqdm
```

```python
def load_X_Y(nb):
    X_fname = "data/Xtr{}.csv".format(nb)
    DNAs = pd.read_csv(X_fname).seq
    X = []
    for DNA in DNAs:
        l_to_id = {"T": 0, "G": 1, "C": 2, "A": 3}
        X.append([l_to_id[l] for l in DNA])
    X = np.array(X)

    Y_fname = "data/Ytr{}.csv".format(nb)
    Y = pd.read_csv(Y_fname).Bound
    Y = np.array(Y)

    return X, Y
```

```python
X0, Y0 = load_X_Y(0)
X1, Y1 = load_X_Y(1)
X2, Y2 = load_X_Y(2)
X= np.concatenate((X0, X1, X2), axis=0)
Y= np.concatenate((Y0, Y1, Y2), axis=0)
n_samples = X.shape[0]
trainprop = 0.75
rand_ind = np.random.permutation(n_samples)
ind_train = rand_ind[:int(trainprop*n_samples)]
ind_val = rand_ind[int(trainprop*n_samples)+1:]
X_train = X[ind_train]
Y_train = Y[ind_train]
X_val = X[ind_val]
Y_val = Y[ind_val]
```

```python
def to_k_info(inp, k):
    p = inp.shape[0]
    inp_new = np.zeros((p-k+1,k))
    for b in range(p-k+1):
        for c in range(k):
            inp_new[b, c] = inp[b+c]
    return inp_new

def to_features(inp_k_info):
    ret = {}
    for f in inp_k_info:
        f_ = tuple(f.tolist())
        if f_ in ret:
            ret[f_] += 1
        else:
            ret[f_] = 1
    return ret

k=3

def my_kernel(inp1, inp2):
    inp1_new = to_features(to_k_info(inp1, k))
    inp2_new = to_features(to_k_info(inp2, k))
    ret = 0
    for f in inp1_new:
        if f in inp2_new:
            ret += inp1_new[f]*inp2_new[f]
    return ret

def my_kernel_matrix(X, Y):
    ret = np.zeros((X.shape[0], Y.shape[0]))
    new_X = []
    new_Y = []
    for i in tqdm(range(X.shape[0])):
        new_X.append(to_features(to_k_info(X[i], k)))
    for i in tqdm(range(Y.shape[0])):
        new_Y.append(to_features(to_k_info(Y[i], k)))
    for i in tqdm(range(X.shape[0])):
        for j in range(Y.shape[0]):
            for f in new_X[i]:
                if f in new_Y[j]:
                    ret[i, j] += new_X[i][f]*new_Y[j][f]
    print("Kernel Matrix Computed")
    return ret
```

```python
my_kernel(X[1], X[2])
```

```python
h = .02  # step size in the mesh

# we create an instance of SVM and fit out data.
clf = svm.SVC(kernel=my_kernel_matrix)
clf.fit(X, Y)
```
